```python
#html parser
from bs4 import BeautifulSoup

def parser(response_text):
    soup = BeautifulSoup(response_text, features="html.parser")
    matches = soup.body.pre.text.lower().split('\n')[1:]

    match = set(matches[0])
    for i in matches:
        match = match & set(i)
        if len(match) == 1:
            break;
    if len(match) == 1:
        return list(match)[0]
    return None
```
